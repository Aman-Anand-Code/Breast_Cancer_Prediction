```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Breast_cancer_data.csv')
df.head(5)
```

```python
df.describe()
```

### Prepraing Data

```python
# Separate features and target variable
x = df.drop('diagnosis', axis=1)
y = df['diagnosis']

# Split the data into training and testing sets
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=0.2, random_state=42)

# Display the shape of the training and testing sets
print(f'Training feature shape: {train_x.shape}')
print(f'Testing feature shape: {test_x.shape}')
print(f'Training labels shape: {train_y.shape}')
print(f'Testing labels shape: {test_y.shape}')
```

### Classification

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

clf_log = LogisticRegression()
clf_log.fit(train_x, train_y)

clf_log.predict(test_x)
```

#### Linear SVM

```python
from sklearn import svm

clf_svm = svm.SVC(kernel='linear')
clf_svm.fit(train_x, train_y)

clf_svm.predict(test_x)
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

clf_dec = DecisionTreeClassifier()
clf_dec.fit(train_x, train_y)

clf_dec.predict(test_x)
```

#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

clf_gnb = GaussianNB()
clf_gnb.fit(train_x, train_y)

clf_gnb.predict(test_x)
```

### Evaluation

```python
# Mean Accuracy
print(f'Logistic Regression Mean Accuracy: {clf_log.score(test_x, test_y)}')
print(f'Linear SVM Mean Accuracy: {clf_svm.score(test_x, test_y)}')
print(f'Decision Tree Mean Accuracy: {clf_dec.score(test_x, test_y)}')
print(f'Naive Bayes Mean Accuracy: {clf_gnb.score(test_x, test_y)}')
```

```python
# F1 Score
from sklearn.metrics import f1_score

print(f'Logistic Regression F1 Score: {f1_score(test_y, clf_log.predict(test_x), average=None, labels=[0, 1])}')
print(f'Linear SVM F1 Score: {f1_score(test_y, clf_svm.predict(test_x), average=None, labels=[0, 1])}')
print(f'Decision Tree F1 Score: {f1_score(test_y, clf_dec.predict(test_x), average=None, labels=[0, 1])}')
print(f'Naive Bayes F1 Score: {f1_score(test_y, clf_gnb.predict(test_x), average=None, labels=[0, 1])}')
```

### Tuning SVM with Grid Search

#### Grid search help us increase the accuracy of our SVC model by selecting the best parameters for our data set from the passed parameter

```python
# Previous Parameter
print(clf_svm.get_params())
```

```python
# Applying Grid Search
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
prameters = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
}

# Create the SVC model
svc = svm.SVC()

# Create the GridSearchCV object
clf = GridSearchCV(svc, prameters, cv=5)
clf.fit(train_x, train_y)

# Print the best parameters and the best score
# print(f'Best Parameters: {grid_search.best_params_}')
# print(f'Best Score: {grid_search.best_score_}')
```

```python
print(f'Best new Parameters: {clf.get_params()}')
print(f'Best Score: {clf.score(test_x, test_y)}')
```

### Saving Model

```python
import pickle

with open('Model_Cancer.pkl', 'wb') as f:
    pickle.dump(clf, f)
```
